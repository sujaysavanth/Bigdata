```python
import pandas as pd

googleplaydf=pd.read_csv('/content/googleplaystore.csv')

googleplaydf
```

```python
# Problem 1: Overall Average Rating (2 points)
# We would like to compute the average rating for all the apps in the dataset. This will give us an idea of which apps are performing above average and which are performing below average, in terms of rating.

# Write a python script that does that. Use Pandas Data Frames with the mean() function applied to the rating column.

# Here are the steps that you should follow in your code:

# Read the csv file into a dataframe using the pandas.read_csv() function
# Extract the 'Rating' column and compute its mean
# Print the mean

appsAverageRating = googleplaydf['Rating'].mean()

print("Average Rating of applications",appsAverageRating)
```

```python
#  Average Rating per Category (3 points)
averageratingpercategory = googleplaydf.groupby('Category')['Rating'].mean()

print("Average Rating per Category")
print(averageratingpercategory)
```

```python
# using for loop
uniquecatergories = googleplaydf['Category'].unique()
print(uniquecatergories)
print("Average ratings per category:")
for category in uniquecatergories:
  percategorydata = googleplaydf[googleplaydf['Category']==category]
  averagerating = percategorydata['Rating'].mean()
  print(f"{category}: {averagerating}")
```

```python
# Non numeric columns
nonnumericdata = googleplaydf.select_dtypes(exclude='number')

print("Non-numeric columns:")
print(nonnumericdata.columns)

numericdata = googleplaydf.select_dtypes(include='number')

print("Numeric columns:")
print(numericdata.columns)
```

```python
nonnumericcolumnlist = nonnumericdata.columns.tolist()

for column in nonnumericcolumnlist:
  uniquelabels = googleplaydf[column].nunique()
  print(f"No of Unique labels in column '{column}':")
  print(uniquelabels)
  valuecounts = googleplaydf[column].value_counts()
  print(f"frequencies of label")
  print(valuecounts)
  print()
```

```python
numericcolumnlist = numericdata.columns.tolist()

for column in numericcolumnlist:
  minvalue = googleplaydf[column].min()
  maxvalue = googleplaydf[column].max()
  meanvalue = googleplaydf[column].mean()
  print(f"Minimum value of column '{column}': {minvalue}")
  print(f"Maximum value of column '{column}': {maxvalue}")
  print(f"Mean value of column '{column}': {meanvalue}")
  print()
```

```python
salesdf = pd.read_csv('/content/sales.csv')

salesdf
```

```python
purchasesperproducttype = salesdf['product_type'].value_counts()
print(purchasesperproducttype)
```

```python
highestpurchasetype = purchasesperproducttype.idxmax()
highestpurchases = purchasesperproducttype.max()
print(f"Highest Purchase Type: {highestpurchasetype} and number of purchases are: {highestpurchases}")
```

```python
salesdf['purchase_date'] = pd.to_datetime(salesdf['purchase_date'])

salesdf['month'] = salesdf['purchase_date'].dt.month_name()

salespermonth = salesdf['month'].value_counts()
print(salespermonth)
```
